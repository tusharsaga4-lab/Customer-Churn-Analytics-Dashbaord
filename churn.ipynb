```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_excel("Customer_Churn_Dataset.xlsx")

# ==========================
# BASIC INFO
# ==========================

print(df.shape)
print(df.head())
print(df.info())

# ==========================
# REMOVE DUPLICATES
# ==========================

df.drop_duplicates(inplace=True)

# ==========================
# HANDLE MISSING VALUES
# ==========================

df.replace(" ", np.nan,)

df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

df["TotalCharges"].fillna(
    df["TotalCharges"].median(),
    
)

# ==========================
# STANDARDIZE TEXT
# ==========================

text_cols = df.select_dtypes(include="object").columns

for col in text_cols:
    df[col] = df[col].str.strip()

# ==========================
# CHURN MAPPING
# ==========================

df["Churn_Flag"] = df["Churn"].map({
    "No":0,
    "Yes":1
})

# ==========================
# TENURE GROUP
# ==========================

bins = [0,12,24,36,48,60,72]

labels = [
    "<1 Year",
    "<2 Year",
    "<3 Year",
    "<4 Year",
    "<5 Year",
    "<6 Year"
]

df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=bins,
    labels=labels
)

# ==========================
# MONTHLY CHARGE SEGMENT
# ==========================

df["ChargeSegment"] = pd.qcut(
    df["MonthlyCharges"],
    4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Premium"
    ]
)

# ==========================
# REVENUE COLUMN
# ==========================

df["Revenue"] = (
    df["MonthlyCharges"] *
    df["tenure"]
)

# ==========================
# SAVE CLEAN DATA
# ==========================

df.to_csv(
    "customer_churn_cleaned.csv",
    index=False
)

print("Data Cleaning Complete")
```
